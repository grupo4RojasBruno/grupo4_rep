## Enunciado
Diseñe un algoritmo y programapara calcular el área de un círculo dado su radio.

## Propuesta respuesta
Para resilver este ejercicio se define una función con la ecuación del área del circulo donde la entrada debe ser el radio de éste mismo en [cm].

```python
# Definición de función

def area_circulo(r):
    pi = 3.14159
    a  = pi*r**2
    
    return a
```

```python
# Ejemplo de uso

radio = 3 # cm
area  = area_circulo(radio)

print("El área del circulo de radio {} [cm] es de {} [cm^2]".format(radio, area))
```

# Ejercicio 8

## Enunciado
Asuma que posee dos variables, “varA” y “varB”, las cuales tienen valores asignados, números o strings.Escriba un código en Python que imprima por pantalla unode los siguientes mensajes:

•“string involucrado”, si ambos varA y varB son strings. 

•“mas grande”, si “varA” is mayor que “varB".

•“igual”, si “varA”, es igual a “varB”.

•“mas pequeño”, si “varA” es menor que “varB”.

```python
# Definición de función

def var_tipo(A, B):
    caso0 = (type(A) == type("")) and (type(B) == type(""))
    caso1 = (A >  B) and not(caso0)
    caso2 = (A == B) and not(caso0)
    caso3 = (A <  B) and not(caso0)
    
    if caso0: print("String involucrado")
    if caso1: print("Más grande")
    if caso2: print("Igual")
    if caso3: print("Más pequeño")
    
    return
```

```python
# Ejemplo de uso

var_tipo("Sebastian","Bruna")
var_tipo(4,1)
var_tipo(32,32)
var_tipo(5,12341232)
```

## Problema a desarrollar

Se desea tener un registro de las características generales del clan familiar Monalta, para esto se recurre a la programación orientada a objetos en python, los familiares a considerar son:

```python
# Clase base
class miembro_clan():
    '''Esta clase describe las características de una persona del clan familiar'''
    
    def __init__(self, nombre, edad):
        self.nombre    = nombre
        self.edad      = edad
    
    def descripcion_basica(self):
        desc = ("Nombre   : {} \n".format(self.nombre) +
                "Edad     : {} \n".format(self.edad))
        return print(desc)
    
    def deseo(self, deseo):
        
        print("{} esta trabajando para lograr lo que desea: {} \n".format(self.nombre, deseo))    
```

```python
juan = miembro_clan("Juan Monalta", 23)
juan.descripcion_basica()
juan.deseo("lograr independencia económica")
```

```python
# Subclase 1 
class rama_principal(miembro_clan):
    def __init__(self,nombre, edad, estatus, funcion, posicion_heredro):
        super().__init__(nombre, edad)
        self.estatus      = estatus
        self.funcion      = funcion
        self.pos_heredero = posicion_heredro
    
    def descripcion_real(self):
        desc = ("Estatus        : {} \n".format(self.estatus) +
                "Función        : {} \n".format(self.funcion) +
                "Lugar heredero : {} \n".format(self.pos_heredero))
        return print(desc)
    
    def habilidad_especial(self, habilidad):
        self.habilidad = habilidad
        print("Ahora la habilidad de {} es: {} \n".format(self.nombre, self.habilidad))
    
```

```python
pepe = rama_principal("Pepe Monalta", 23, "Hijo del patriarca", "Encargado de negocios nacionales","Primer heredero",)
pepe.descripcion_basica()
pepe.descripcion_real()
pepe.deseo("Servir mejor a la familia")
pepe.habilidad_especial("Super memoria")
```

```python
# Subclase 2
class rama_secundaria(miembro_clan):
    def __init__(self,nombre, edad, funcion, jefe):
        super().__init__(nombre, edad)
        self.funcion = funcion
        self.jefe    = jefe
    
    def descripcion_patronazgo(self):
        desc = ("Función         : {} \n".format(self.funcion) +
                "Supervisado por : {} \n".format(self.jefe))
        return print(desc)
    
    def nombre_rama(self, rama):
        self.rama = rama
        print("Ahora la rama a la que pertenece {} es: {} \n".format(self.nombre, self.rama))
```

```python
pedro = rama_secundaria("Pedro Monarez", 23, "Encargado de seguridad","Pepe Moralez",)
pedro.descripcion_basica()
pedro.descripcion_patronazgo()
pedro.deseo("Asegurar la integridad familiar")
pedro.nombre_rama("ABranch")
```

```python
# Subclase 3
class externo(rama_secundaria):
    def __init__(self,nombre, edad, funcion, jefe, profesion):
        super().__init__(nombre, edad, funcion, jefe)
        self.profesion = profesion
    
    def antecedentes(self, antecedente_penal, antecedente_laboral):
        self.antecedente_penal    = antecedente_penal
        self. antecedente_laboral = antecedente_laboral
        ant = ("El antecedente penal de {} es {} \n".format(self.nombre, self.antecedente_penal) + 
               "El antecedente laboral de {} es {} \n".format(self.nombre, self.antecedente_laboral))
        print(ant)
        
    def observacion(self, observacion):
        self.observacion = observacion
        obs = ("Las observaciones del {} son: {}".format(self.nombre, self.observacion))
        print(obs)
```

```python
clemente = externo("Clemente Simple", 21, "Cientifico de datos","Pepe Moralez", "Ingeniero")
clemente.descripcion_basica()
clemente.descripcion_patronazgo()
clemente.deseo("No ser despdido y conseguir un aumento de sueldo")
clemente.nombre_rama("CBranch")
clemente.antecedentes("De adolecente rayó una pared", "No posee, este es su primer trabajo")
clemente.observacion("Es un buen trabajador, cuando sea posible aumentarle el sueldo")
```
